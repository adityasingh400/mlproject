```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df_batsman = pd.read_csv('batsman data odi.csv')
numeric_cols =df_batsman .select_dtypes(include='number').columns
sns.pairplot(df_batsman[numeric_cols], diag_kind='hist')
plt.tight_layout()
plt.show()
```

```python
sns.heatmap(df_batsman[numeric_cols].corr(),annot=True, cmap='coolwarm')
plt.show()
```
