```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
# convert x to strings and drop rows with missing x or y to avoid matplotlib type errors
x = df.iloc[:, 1]
y = df.iloc[:, 2]
mask = x.notna() & y.notna()
x = x[mask].astype(str)
y = y[mask]
plt.plot(x, y)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
